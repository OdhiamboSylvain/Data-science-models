DATA READING


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('depressionData.csv')
df.head()
```

```python
print(df.shape)
```

```python
print(df.dtypes)
```

DATA PROCESSNG

```python
print(df.isnull().sum())#checking rows with missing values
```

```python
df.drop(columns = ['Timestamp'], inplace = True) #dropping columns we dont need
```

```python
df.dropna(inplace = True)#dropping rows wiith missing values
print(df.shape)
```

```python
print(df.columns)
```

```python
# encoding the categorical variables
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])


# convert all data to int
df = df.astype('int')

df.head()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler#scaling of features
scaler = StandardScaler()
df[['Age', 'Feeling sad', 'Trouble sleeping at night', 'loss of appetite', 'Irritable towards people', 'Suicide attempt', 'Problems of bonding with people', 'Feeling of guilt', 'Problems concentrating or making decision']] = scaler.fit_transform(df[['Age', 'Feeling sad', 'Trouble sleeping at night', 'loss of appetite', 'Irritable towards people', 'Suicide attempt', 'Problems of bonding with people', 'Feeling of guilt', 'Problems concentrating or making decision']])
df.describe()
```

```python
# Separating the target variable and the independent variable
y = df.pop("Depressed")
X = df
```

```python
# Shuffling the data

df = df.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
# Randomization

# function to randomly return a value between 0 and 1
import random

ts = (random.randint(5, 50)) / 100
print(ts)

rs = random.randint(1, 50)
print(rs)
```

```python
from sklearn.model_selection import train_test_split




X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=ts, random_state=rs)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

MODELLING

```python
from sklearn.neighbors import KNeighborsClassifier          # we are going to classify weather a person is depressed or not
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# instantiate the models
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()

# fit the models
knn.fit(X_train, y_train)
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
# predict the values
y_pred_knn = knn.predict(X_test)
y_pred_dt = dt.predict(X_test)
y_pred_rf = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print('KNN')
print(confusion_matrix(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
print('Accuracy Score is:' , accuracy_score(y_test, y_pred_knn))
print('---' * 20)

print('DT')
print(confusion_matrix(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))
print('Accuracy Score is:' , accuracy_score(y_test, y_pred_dt))
print('---' * 20)

print('RF')
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
print('Accuracy Score is:' , accuracy_score(y_test, y_pred_rf))
print('---' * 20)
```
